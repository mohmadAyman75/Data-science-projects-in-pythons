#### I will use the .describe() function to identify which columns need to be scaled down in order to make it easier for the model to perform statistical operations.

```python
df['sqft_ratio'] = df['sqft_living'] / df['sqft_lot']
df['rooms_per_floor'] = df['bedrooms'] / df['floors']
df['bath_per_bedroom'] = df['bathrooms'] / df['bedrooms']
```

```python
df.describe()
```

#### ["price", "sqft_lot"] need to be scaled using either StandardScaler or MinMaxScaler.

```python
from sklearn.preprocessing import StandardScaler
```

```python
std=StandardScaler()
```

```python
col_std=["price","sqft_lot","sqft_ratio"]
X = df[col_std]
data_Scaled=std.fit_transform(X)
df = df.reset_index(drop=True)
new_data_scaled = pd.DataFrame(data_Scaled, columns=col_std)
```

```python
new_data_scaled.describe()
```

```python
final_df = pd.concat([df.drop(columns=col_std), new_data_scaled], axis=1)
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.describe()
```

```python
final_df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=final_df.drop("price",axis=1)
Y=final_df["price"]
X = X.replace([np.inf, -np.inf], np.nan)

X = X.dropna()

Y = Y.loc[X.index]
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
```

```python
LR=LinearRegression()
RG=Ridge()
LA=Lasso()
```

```python
LR.fit(x_train,y_train)
RG.fit(x_train,y_train)
LA.fit(x_train,y_train)
```

```python
LR_predict=LR.predict(x_test)
RG_predict=RG.predict(x_test)
LA_predict=LA.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
LR_r2_score=r2_score(y_test,LR_predict)
RG_r2_score=r2_score(y_test,RG_predict)
LA_r2_score=r2_score(y_test,LA_predict)
print(f"this is accuracy in linear regression:{LR_r2_score}")
print(f"this is accuracy in ridge:{RG_r2_score}")
print(f"this is accuracy in lasso:{LA_r2_score}")
```

#### to solve the problem faced us we will use a strong model like ("RandomForestRegressor")

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=300,
    max_depth=20,
    min_samples_split=5,
    random_state=42
)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
r2_rf = r2_score(y_test, y_pred)
print("Linear Regression R² Score:", LR_r2_score)
print("Random Forest R² Score:", r2_rf)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.scatter(y_test, LR_predict, alpha=0.5)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         'r--')
plt.title("Actual vs Predicted House Prices - Linear Regression")
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.subplot(1,2,2)
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted House Prices - Random Forest")
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         'r--')  # perfect prediction line
plt.tight_layout() 
plt.show()
```

In this experiment, with just a very small adjustment to the dataset (I only added three columns), the model’s performance improved by about 2%.
Then, when I used a stronger model, the accuracy increased by nearly 10%.

This clearly shows that the problem isn’t always about removing outliers or blaming the dataset — sometimes, the real issue lies in choosing the right model.

And that’s the beauty of machine learning — nothing is ever fixed or absolute.
