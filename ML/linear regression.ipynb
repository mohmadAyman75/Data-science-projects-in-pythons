#### Although the model’s accuracy is quite low due to the presence of many outliers, the main issue is not actually the outliers themselves, and I will demonstrate this now .
i will remove outlietrs and train model again 

```python
numeric_cols = final_df.select_dtypes(include=["int64", "float64"]).columns

for col in numeric_cols:
    Q1 = final_df[col].quantile(0.25)
    Q3 = final_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    final_df = final_df[(final_df[col] >= lower_bound) & (final_df[col] <= upper_bound)]
```

```python
X=final_df.drop("price",axis=1)
Y=final_df["price"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
Try_two_LR=LinearRegression()
Try_two_RG=Ridge()
Try_two_LA=Lasso()
```

```python
Try_two_LR.fit(x_train,y_train)
Try_two_RG.fit(x_train,y_train)
Try_two_LA.fit(x_train,y_train)
```

```python
LR_predict=Try_two_LR.predict(x_test)
RG_predict=Try_two_RG.predict(x_test)
LA_predict=Try_two_LA.predict(x_test)
```

```python
from sklearn.metrics import r2_score
LR_r2_score=r2_score(y_test,LR_predict)
RG_r2_score=r2_score(y_test,RG_predict)
LA_r2_score=r2_score(y_test,LA_predict)
print(f"this is accuracy in linear regression:{LR_r2_score}")
print(f"this is accuracy in ridge:{RG_r2_score}")
print(f"this is accuracy in lasso:{LA_r2_score}")
```

#### See?! This happens because the model we used is a simple one, while the dataset is large and contains important outliers that we cannot remove since they represent real data.
##### The solution to this issue is to try a more powerful model that can handle such data.
